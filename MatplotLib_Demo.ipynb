# Matplot_Example

```python

years=[2011,2012,2013,2014,2015]
yield_apples=[0.895,0.915,0.80,0.935,0.945]

plt.plot(years,yield_apples)
```

```python
plt.plot(years,yield_apples)
plt.xlabel("Year")   #x axis
plt.ylabel("yield(tons)")  #y axis
```

```python
years=range(2000,2008)
yield_apples=[.895, .915, .920, .935, .945, .953, .962, .967]
yield_orange=[.923, .927, .933, .937, .942, .948, .955, .961]

plt.figure(figsize=(8,6))
plt.plot(years,yield_apples)
plt.plot(years,yield_orange)

plt.xlabel("Year")
plt.ylabel("Yield (tons)")
plt.title("Crop yield in state")
plt.legend(["Apples", "Oranges"])
plt.show()
```

```python
#Scatter help to entimate the specific point for 2 diff. data feature relationship

plt.scatter(years,yield_apples)
plt.scatter(years,yield_orange)

plt.xlabel("Year")
plt.ylabel("Yield")

plt.legend(["Apples", "Oranges"])
```

```python
#Create Subplot

names=['A','B','C']
values=[1,10,100]

plt.figure(figsize=(9,3))

plt.subplot(131)
plt.bar(names,values)
plt.subplot(132)
plt.bar(names,values)
plt.subplot(133)
plt.bar(names,values)

plt.suptitle("Categorical Plotting")        # .suptitle()---->Help to display the column name of all the subplots
plt.show()
```
