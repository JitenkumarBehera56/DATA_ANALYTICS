```python
numeric_data = titanic.select_dtypes(include=['number'])        




correlation_matrix = titanic.corr(numeric_only=True)    # Include only numeric columns (like integers and floats)    
                                                        # Correlation only works with numbers  not with text like "male", "female" or "S","C".
                                                        # .corr()---> It Calculates pairwise correlation between columns in a DataFrame.

correlation_matrix
```

```python
# Automatically select desired columns: only numeric or boolean (True/False) columns
titanic_subset = titanic.select_dtypes(include=['number', 'bool']).dropna()

# Plot the heatmap directly
plt.figure(figsize=(10,5))

#It creates a heatmap (colored grid chart) that visually displays the correlation between all numeric and boolean columns in the titanic_subset DataFrame.
sns.heatmap(titanic_subset.corr(), cbar=True, annot=True, fmt=".6f", linewidth=2, cmap="Blues")

plt.title("Correlation Heatmap - Titanic (Numeric & Boolean Columns)")
plt.show()




#sns.heatmap(...) ------> Calls Seaborns heatmap function to create the chart.
#titanic_subset.corr() ------> Calculates how strongly each pair of numeric/boolean columns are correlated (e.g. survived vs. age).
#cbar=True ------> Adds a color bar (legend) that shows what the colors mean.
#fmt=".6f"------> Formats those numbers to 6 decimal places (e.g., 0.812500).
#linewidth=2------> Adds thick white lines between the cells.
#cmap="Blues"------> Uses different shades of blue to represent correlation strength.

```
